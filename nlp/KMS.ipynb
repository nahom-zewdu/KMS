```python
import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

# Sample Slack message
text = "Alice redesigned the payment API because of Jira #456."

# Process text
doc = nlp(text)

# Extract tokens and part-of-speech (POS) tags
print("Tokens and POS:")
for token in doc:
    print(f"{token.text}: {token.pos_}")

# Extract named entities
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```
